```python
import pandas as pd
```

```python
# Read csv
weather_data="weather.csv"
weather_data = pd.read_csv(weather_data)
weather_data
```

```python
# Drop city_name, lat, lon, NaN values, weather_id, weather_icon
# weather_data.drop(columns = ['city_name', 'lat','lon','weather_id','weather_icon'])

weather_data = weather_data.dropna(axis='columns')
weather_data["Date"] = (weather_data["dt_iso"].str.split(" ", n = 1, expand = True))[0]
weather_data["Hour"] = (weather_data["dt_iso"].str.split(" ", n = 1, expand = True))[1].str.split(":", n =1 , expand = True)[0]
weather_data = weather_data.drop(['dt_iso'],axis=1)
weather_data
```

```python
# Filter 01-05-2003 to 11-07-2020
weather_data = weather_data[(weather_data['dt'] >= 1041742800) & (weather_data['dt'] <= 1604725200)]
weather_data
```

```python
weather_data.to_csv("weather_data.csv")
```
