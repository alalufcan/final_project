```python
import requests
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

### HYDRO DATA

```python
# Reading Hydro Data
engine  = create_engine('postgresql://postgres:postgres@dbname.cxw2xnixkpbl.ca-central-1.rds.amazonaws.com/postgres')

# Reflect an existing database into a new model
Base = automap_base()

# Reflect the tables
Base.prepare(engine, reflect=True)

# Save reference to the table
hydro = Base.classes.hydro
weather = Base.classes.weather
```

```python
session = Session(engine)
results = session.query(hydro.date,hydro.hour,hydro.demanded_toronto,hydro.weekday).all()

session.close()
    
hydro = []
for date, hour, demanded_toronto, weekday in results:
    hydro_dict = {}
    hydro_dict["date"] = date
    hydro_dict["hour"] = hour
    hydro_dict["demanded_toronto"] = demanded_toronto
    hydro_dict["weekday"] = weekday
    hydro.append(hydro_dict)

hydro[:5]
```

```python
Date = []
Hour = []
Demanded_Toronto = []
Weekday = []

for i in range (len(hydro)):
    Date.append(hydro[i]['date'])
    Hour.append(hydro[i]['hour'])
    Demanded_Toronto.append(hydro[i]['demanded_toronto'])
    Weekday.append(hydro[i]['weekday'])

    
hydro_df = pd.DataFrame({'Date': Date,
                           'Hour': Hour,
                           'Demanded_Toronto': Demanded_Toronto,
                           'Weekday': Weekday})
```

### WEATHER DATA

```python
session = Session(engine)
results = session.query(weather.date,weather.hour,weather.dt,weather.timezone,weather.temp,weather.feels_like,weather.temp_min,weather.temp_max,weather.pressure,weather.humidity,weather.wind_speed,weather.wind_deg,weather.clouds_all,weather.weather_main,weather.weather_description).all()

session.close()
       
weather = []

for date,hour,dt,timezone,temp,feels_like,temp_min,temp_max,pressure,humidity,wind_speed,wind_deg,clouds_all,weather_main,weather_description in results:
    weather_dict = {}
    weather_dict["date"] = date
    weather_dict["hour"] = hour
    weather_dict["dt"] = dt
    weather_dict["timezone"] = timezone
    weather_dict["temp"] = temp
    weather_dict["feels_like"] = feels_like
    weather_dict["temp_min"] = temp_min
    weather_dict["temp_max"] = temp_max
    weather_dict["pressure"] = pressure
    weather_dict["humidity"] = humidity
    weather_dict["wind_speed"] = wind_speed
    weather_dict["wind_deg"] = wind_deg
    weather_dict["clouds_all"] = clouds_all
    weather_dict["weather_main"] = weather_main
    weather_dict["weather_description"] = weather_description
    weather.append(weather_dict)



weather[:5]




```

```python
Date = []
Hour = []
Clouds = []
UTC_Date =[]
Feels_Like = []
Humidity = []
Pressure = []
Temperature = []
Temp_Max = []
Temp_Min = []
TimeZone = []
Weather_Description = []
Weather_Main = []
Wind_Direction = []
Wind_Speed = []

for i in range (len(weather)):
    Date.append(weather[i]['date'])
    Hour.append(weather[i]['hour'])
    Clouds.append(weather[i]['clouds_all'])
    UTC_Date.append(weather[i]['dt'])
    Feels_Like.append(weather[i]['feels_like'])
    Humidity.append(weather[i]['humidity'])
    Pressure.append(weather[i]['pressure'])
    Temperature.append(weather[i]['temp'])
    Temp_Max.append(weather[i]['temp_max'])
    Temp_Min.append(weather[i]['temp_min'])
    TimeZone.append(weather[i]['timezone'])
    Weather_Description.append(weather[i]['weather_description'])
    Weather_Main.append(weather[i]['weather_main'])
    Wind_Direction.append(weather[i]['wind_deg'])
    Wind_Speed.append(weather[i]['wind_speed'])
    
    
weather_df = pd.DataFrame({'Date': Date,
                           'Hour': Hour,
                           'Clouds': Clouds,
                           'UTC_Date': UTC_Date,
                           'Feels_Like': Feels_Like,
                           'Humidity': Humidity,                          
                           'Pressure': Pressure,                          
                           'Temperature': Temperature,
                           'Temp_Max': Temp_Max,
                           'Temp_Min': Temp_Min,
                           'TimeZone': TimeZone,
                           'Weather_Description': Weather_Description,
                           'Weather_Main': Weather_Main,
                           'Wind_Direction': Wind_Direction,
                           'Wind_Speed': Wind_Speed                          
                          })
```

```python
# Merge weather and hydro datasets 
combined_df = pd.merge(weather_df, hydro_df, on= ['Date', 'Hour'], how="inner") 

# Drop unneccesary columns

# combined_df = combined_df.drop(['UTC_Date', 'TimeZone', 'Weather_Description', 'Weather_Main', 'Feels_Like', 'Wind_Direction', 'Clouds', 'Temp_Max', 'Temp_Min'], axis =1)
# combined_df = combined_df.drop(['UTC_Date', 'TimeZone', 'Weather_Description', 'Weather_Main', 'Feels_Like', 'Wind_Direction', 'Clouds'], axis =1)
# combined_df = combined_df.drop(['UTC_Date', 'TimeZone', 'Weather_Description', 'Weather_Main', 'Feels_Like', 'Wind_Direction'], axis =1)
combined_df = combined_df.drop(['Date', 'UTC_Date', 'TimeZone', 'Weather_Description', 'Weather_Main', 'Feels_Like', 'Wind_Direction', 'Pressure', 'Temp_Max', 'Temp_Min', 'Clouds'], axis =1)


combined_df.head()
```

```python
# Add demand from previous day as a new feature

combined_df['Previous_Day_Demand'] = ''
combined_df['Previous_Hour_Temperature'] = ''

for i in range(24, len(combined_df)):
  combined_df.loc[i,'Previous_Day_Demand'] = combined_df.loc[i-24, 'Demanded_Toronto']
  combined_df.loc[i-23,'Previous_Hour_Temperature'] = combined_df.loc[i-23, 'Temperature'] - combined_df.loc[i-24, 'Temperature']

for i in range(23):
  combined_df.loc[i,'Previous_Day_Demand'] = combined_df.loc[i,'Demanded_Toronto']

combined_df.loc[0,'Previous_Hour_Temperature'] = combined_df.loc[0, 'Temperature']
combined_df.head()
```
